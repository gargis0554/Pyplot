```python
cities=["chennai","mumbai","delhi"]
```

```python
val=[[5.,10.,15.,20.],[4.,8.,12.,16.],[6.,12.,18.,24.]]
```

# PIE CHART

```python
arg=[1.25,8.8,9,14]
```

```python
pl.pie(arg)
```

```python
houses=["vidya","ankita","tanya","karuna"]
```

```python
pl.pie(arg,labels=houses,autopct="%05.3f%%")
```

```python
colr=["blue","green","red","yellow"]
```

```python
pl.pie(arg,colors=colr)
```

```python
expl=[0,0.2,0.3,0]
```

```python
pl.pie(arg,explode=expl)
```

```python
pl.title("My pie chart")
pl.pie(arg)
```

```python
lb=["1","2","3","4"]
```

```python
pl.pie(arg,labels=lb)
pl.legend(loc='upper left')
```
